```python
'''
MiniProject2 - Scraping 50 projects off of SourceForge
All projects start with assigned letter, 'z'
'''
import requests
from bs4 import BeautifulSoup
import re
import pymongo
import itertools

dbname = "fdac18mp2" #please use this database
collname = "glprj_akarnauc" #please modify so you store data in your collection
client = pymongo.MongoClient()

db = client[dbname]
coll = db[collname]

letter = 'z'
final = set()

def find_projects(text):
    matches = re.findall(r'(?<=\/projects\/)(.*?)\/', text.lower())
    letter_matches = set([proj for proj in matches if proj[0] == letter])
    return letter_matches
    
def get_projs(URL):
    global final
    base = 'https://sourceforge.net/projects/'
    r = requests.get(URL)
    projs = find_projects(r.text)
    urls = [base + str(proj) + '/' for proj in projs]
    final = final.union(urls)

index = 23
while len(final) < 50:
    base = 'https://sourceforge.net/directory/?q=z&sort=name&page=' + str(index)
    get_projs(base)
    index += 1

for i, url in enumerate(itertools.islice(final, 50)):
    print({"url": url})
    coll.insert({"url": url})
```
